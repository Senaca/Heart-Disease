* *Veri Setinde bulunan sütunlar ve açıklamaları;*
* 1.age **(Yaş)**
1. 2.sex(cinsiyet)**(1 = Erkek, 0 = Kadın)**
1. 3.cp:chest pain type **(göğüs ağrısı türü) (3: atipik anjin, 2: tipik anjina, 1: anjinal olmayan ağrı, 0: asemptomatik)**
1. 4.trestbps: resting blood pressure(in mm Hg on admission to the hospital- ) **Kişinin istirahat kan basıncı**
1. 5.chol: serum cholestoral in mg/dl **(Serum Kollestrolü)**
1. 6.fbs: (fasting blood sugar > 120 mg/dl)(1 = true; 0 = false) **Kişinin açlık kan şekeri (> 120 mg / dl, 1 = doğru; 0 = yanlış)**
1. 7.restecg: resting electrocardiographic results (values 0,1,2) **(Elektrokardiografik sonuçların dinlenmesi)**
1. 8.thalach: maximum hear rate achieved **(maksimum kalp atış hızı)**
1. 9.exang: exercise induced angina *(1 = yes; 0 = no)* **Egzersize bağlı anjin**
1. 10.oldpeak: ST depression induced by exercise relative to test **(Teste göre indüklenen depresyon)**
1. 11.slope: the slope of the peak exercise ST segment **(Tepe egzersizi ST segmentinin eğimi)** *(1:yukarı, 2:düz, 3:aşağı yönlü)*
1. 12.ca: number of major vessels(0-3) colored by flourosopy **(floroskopi ile renklendirilmiş ana damarların (0-3) sayısı)**
1. 13.thal: 1 = normal; 2 = fixed defect; 7 = reversable defect **Talasemi (1 = normal; 2 = sabit kusur; 3 = tersinir kusur)**
1. 14.target: 1 or 0 **Kalp hastalığı**

```python
ratio = []
for i in thal_list:
    x = df[df['thal']==i]
    rate = sum(x.trestbps)/len(x)  
    ratio.append(rate)    
data = pd.DataFrame({'thal_list':thal_list ,'ratio':ratio}) 
new_index = (data['ratio'].sort_values(ascending=False)).index.values
sorted_data = data.reindex(new_index)

# visualization
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['thal_list'], y=sorted_data['ratio'])
plt.xticks(rotation= 45)
plt.xlabel('Thal')
plt.ylabel('TrestbpsRate')
plt.title('Rate')
```
